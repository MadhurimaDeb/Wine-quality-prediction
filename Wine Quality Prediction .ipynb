```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from sklearn.datasets import load_wine 
```

```python
Df = load_wine()
```

```python
Df
```

```python
df = pd.DataFrame(Df.data , columns = Df.feature_names)
```

```python
df
```

```python
df.head()
```

```python
df['target'] = Df.target
```

```python
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
all_columns = df.columns
```

```python
# Perform outlier analysis on numerical variables
for var in all_columns:
    # Calculate z-scores
    z_scores = (df[var] - df[var].mean()) / df[var].std()
    # Identify outliers using a threshold (e.g., z-score > 3 or < -3)
    outliers = df[z_scores.abs() > 3]
    if not outliers.empty:
        print(f'Outliers in {var}:')
        print(outliers)
```

```python
df.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Scale the numeric variables
df[all_columns] = scaler.fit_transform(df[all_columns])

# Print the scaled data
print(df[all_columns])
```

```python
df.describe()
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(df.corr(), cbar = True, square = True , fmt = '.1f' , annot = True , annot_kws = {'size' : 8} , cmap = 'Blues')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target'] , axis = 1)
y = df['target']
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.1, random_state = 10)
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
scalar.fit(X_train)
X_train = scalar.fit_transform(X_train)
X_test = scalar.fit_transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.metrics import r2_score 
from sklearn.linear_model import LinearRegression
lr_model=LinearRegression()
lr_model.fit(X_train,y_train)
y_pred=lr_model.predict(X_test)
score1 = r2_score(y_test,y_pred)
print(score1)
```

```python
from sklearn.ensemble import RandomForestRegressor
RF_Model=RandomForestRegressor(n_estimators=650,random_state=245,min_samples_leaf=.0001)
RF_Model.fit(X_train,y_train)
y_predict=RF_Model.predict(X_test)
score2 = r2_score(y_test,y_predict)
print(score2)
```

```python
from sklearn.ensemble import  ExtraTreesRegressor
ET_Model=ExtraTreesRegressor(n_estimators = 120)
ET_Model.fit(X_train,y_train)
y_predict=ET_Model.predict(X_test)
score3 = r2_score(y_test,y_predict)
print(score3)
```

```python
from xgboost import XGBRegressor
xgb=XGBRegressor()
xgb.fit(X_train,y_train)
y_predict=xgb.predict(X_test)
score3 = r2_score(y_test,y_predict)
print(score3)
```

```python
from sklearn.model_selection import cross_val_score

mse = cross_val_score(lr_model,X_train,y_train, scoring='neg_mean_squared_error' , cv=5)
mean_mse = np.mean(mse)
print(mean_mse)
```

```python
from sklearn.model_selection import GridSearchCV

params = {'alpha':[1e-15, 1e-10, 1e-8, 1e-4, 1e-3, 1e-2, 1, 5, 10, 20]}
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge = Ridge()


ridge_regressor = GridSearchCV(ridge, params , scoring = 'neg_mean_squared_error', cv=5,n_jobs=-1)
ridge.fit(X_train,y_train)
ridge.score(X_train,y_train)
```

```python
ridge_regressor.fit(X_train,y_train)
```

```python
print(ridge_regressor.best_params_)
```

```python
print(ridge_regressor.best_score_)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
las = Lasso()

las_regressor = GridSearchCV(las, params, scoring = 'neg_mean_squared_error', cv=5)

las_regressor.fit(X_train,y_train)

print(las_regressor.best_params_)
print(las_regressor.best_score_)
```

```python
y_pred = las_regressor.predict(X_test)
print(y_pred)
```

```python
r2_score = r2_score(y_test,y_pred)
print(r2_score)
```

```python
predicted_data = las_regressor.predict(X_train)
```

```python
plt.scatter(y_train,predicted_data)
plt.xlabel("Actual price")
plt.ylabel("Predicted price")
plt.title("Actual price vs Predicted Price")
plt.show()
```
